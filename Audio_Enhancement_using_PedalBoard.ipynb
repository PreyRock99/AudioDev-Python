# **Audio Denoising and Enhancement using Python**
In this demonstration, we will learn how to denoise and enhance an audio file using various Python libraries. We will utilize pedalboard for audio processing, noisereduce for noise reduction, and librosa for loading and playing audio files. This step-by-step guide will show you how to effectively clean up and improve the quality of your audio recordings.

***Step 1:*** Import Libraries and Load Audio File
First, we import the necessary libraries and load the audio file that we want to process.

***Step 2***: Reduce Noise in the Audio File
Using the noisereduce library, we will apply noise reduction to the loaded audio file. This step will help us remove unwanted background noise.

***Step 3***: Apply Audio Effects
Next, we will apply a series of audio effects using the pedalboard library. These effects include a noise gate, compressor, low-shelf filter, and gain adjustment to enhance the audio quality.

```python
from pedalboard.io import AudioFile
from pedalboard import *
import noisereduce as nr
import soundfile as sf
sr=44100
with AudioFile('/content/morgan-freeman-convert-file.wav').resampled_to(sr) as f:
    audio = f.read(f.frames)

reduced_noise = nr.reduce_noise(y=audio, sr=sr, stationary=True, prop_decrease=0.75)

board = Pedalboard([
    NoiseGate(threshold_db=-30, ratio=1.5, release_ms=250),
    Compressor(threshold_db=-16, ratio=4),
    LowShelfFilter(cutoff_frequency_hz=400, gain_db=10, q=1),
    Gain(gain_db=2)
])

effected = board(reduced_noise, sr)


with AudioFile('/content/audio1_enhenced.wav', 'w', sr, effected.shape[0]) as f:
  f.write(effected)

```
