# Load the GTZAN Dataset
In this step, we will load the GTZAN dataset using the datasets library from Hugging Face. The GTZAN dataset is a popular dataset for music genre classification and contains a variety of music genres.

```python
from datasets import load_dataset

gtzan = load_dataset("marsyas/gtzan", "all")
gtzan
```

# Split the GTZAN Dataset into Training and Testing Sets
In this step, we will split the GTZAN dataset into training and testing subsets. We will use a seed for reproducibility and shuffle the data to ensure a random distribution. The split will help in training and evaluating models effectively.

```python
gtzan = gtzan["train"].train_test_split(seed=42, shuffle=True, test_size=0.1)
gtzan
```

```python
gtzan["train"][1]
```

# Display an Audio Sample from the GTZAN Dataset
In this step, we will load an audio sample from the test set of the GTZAN dataset and display it using IPython's audio display capabilities. This allows us to listen to the sample directly in the notebook.

```python


import pandas as pd
import IPython.display as ipd


example = gtzan["test"][1]
ipd.Audio(example["audio"]["array"], rate=example["audio"]["sampling_rate"])
```

# Convert Genre IDs to Genre Labels
In this step, we will convert the genre ID of an audio sample to its corresponding genre label using the int2str function from the GTZAN dataset. This allows us to interpret the genre classification in a human-readable format.

```python
id2label_fn = gtzan["test"].features["genre"].int2str
id2label_fn(gtzan["test"][1]["genre"])
```
